## 目标

- 读取、展示、存储视频
- 从camera中捕捉视频并展示
- cv2.VideoCapture(), cv2.VideoWriter()

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0) # 0表示摄像头的索引编号

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    print(ret,frame)
    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

为了捕捉视频，实现需要创建一个`VideoCapture`对象。Its argument can be either the device index or the name of a video file. Device index is just the number to specify which camera. Normally one camera will be connected (as in my case). So I simply pass 0 (or -1). You can select the second camera by passing 1 and so on. After that, you can capture **frame-by-frame**. But at the end, don’t forget to release the capture.

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)
# cap.set(3,640)
# cap.set(4,480)

# 这是在mac上测试通过的方法，可以用于存储捕捉的图像
w = cap.get(cv2.CAP_PROP_FRAME_WIDTH);
h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT); 
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output_flipped.mp4',fourcc, 12.0,(int(w),int(h)))


while True:
	ret,frame = cap.read()
	gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) # 每帧转为灰度图，
	
	if ret == True:
		frame = cv2.flip(frame,0) # 翻转帧后存储
		out.write(frame)

		cv2.imshow('frame',frame)
		cv2.imshow('gray',gray)

	if cv2.waitKey(1) & 0xFF == ord('q'): # & 0xFF是针对64位机器
		break

cap.release() # 释放摄像头
out.release()
cv2.destroyAllWindows()
```
