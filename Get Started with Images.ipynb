````python
import numpy as np
import cv2
````

````python
img = cv2.imread('dog.png',0) # 以灰度图像加载
````

读取当前目录下的文件。

Use the function cv2.imread() to read an image. The image should be in the working directory or a full path of image should be given.

Second argument is a flag which specifies the way image should be read.

- cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
- cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode
- cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel

分别对应1，0，-1。即1为彩色图像，0为灰度图像，-1加载包含alpha通道。

````python
cv2.imshow('image',img) # 窗口大小自动适应图像大小
cv2.waitKey(0)
cv2.destroyAllWindows()
````

imshow函数的第一个参数是窗口的名字，第二个是需要展示的图像对象。
`cv2.destroyWindow(name) `可以销毁指定的窗口。

```
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
先定义窗口的名字和调整的属性，第二个参数默认是`cv2.WINDOW_AUTOSIZE`, 改成`cv2.WINDOW_NORMAL`可以进行大小调整。

## 写图像到文件

````python
cv2.imwrite('res.jpg',img)
````

## 小总结

Below program loads an image in grayscale, displays it, save the image if you press ‘s’ and exit, or simply exit without saving if you press ESC key.

````python
import numpy as np
import cv2 

img = cv2.imread('python.png',0)
cv2.namedWindow('image',cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
k = cv2.waitKey(0) & 0xFF
if k == 27: # ESC
    cv2.destroyAllWindows()
elif k == ord('s'):
    cv2.imwrite('res_python.jpg',img)
    cv2.destroyAllWindows()
````

## 使用Matplotlib

````python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('python.png',0)
plt.imshow(img,cmap='gray',interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
````
